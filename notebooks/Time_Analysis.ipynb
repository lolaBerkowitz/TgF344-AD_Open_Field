## Reshape dataframe so path_length_tim(n) variables are in long format.

```r
library(tidyr)

df_long <- gather(df, time_bin, path_length_time, path_length_tim1:path_length_tim6, factor_key=TRUE)
df_long = df_long[df_long['day'] == 'D1',]
summary(df_long)
```

## Run a mixed model with subject ID as a random variable and group X time_bin as fixed factor. 

```r
library(lme4)
library(lmerTest)

# full model
mld_f = lmer(log10(path_length_time) ~ group*time_bin + (1|subID), data=df_long)
summary(mld_f)

# reduced model
mld_r = lmer(log10(path_length_time) ~ 1 + (1|subID), data=df_long)
summary(mld_r)
```

```r
# Check model residuals
plot(mld_f)
```

```r
## QQ plot
qqnorm(residuals(mld_f))
```

```r
# Compare model with AIC 

library(AICcmodavg)
aictab(cand.set=list(mld_r, mld_f),
       modnames=c("restricted", "full"),
       sort=FALSE, second.ord=FALSE)

anova(mld_r, mld_f)
```

```r
library(multcomp)
contr <- glht(mld_f, linfct=mcp(time_bin="Tukey"))
summary(contr)
confint(contr)
```

Using tukey multiple comparisons we find that the effect of time_bin is driven by differences between the first and latter time bins. Overall, it seems that rats explore less after the first five minutes of exploration. 

```r
data_summary <- function(data, varname, groupnames){
  require(plyr)
  summary_func <- function(x, col){
    c(mean = mean(x[[col]], na.rm=TRUE),
      sd = sd(x[[col]], na.rm=TRUE))
  }
  data_sum<-ddply(data, groupnames, .fun=summary_func,
                  varname)
  data_sum <- rename(data_sum, c("mean" = varname))
 return(data_sum)
}

df2 <- data_summary(df_long, varname="path_length_time", 
                    groupnames=c("time_bin", "group"))
```

```r
library(ggplot2)
p <- ggplot(df2, aes(x=time_bin, y=path_length_time, group=group, color=group)) + 
          geom_line() +
          geom_point()+
          geom_errorbar(aes(ymin=path_length_time-sd, ymax=path_length_time+sd), width=.2,
                         position=position_dodge(0.05))

p + labs(title="Path Length across time", x="Time (5min/bin)", y = "Path Length(cm)")+
     theme_classic() +
     scale_color_manual(values=c('#601a4a','#9c9eb5'))
```
